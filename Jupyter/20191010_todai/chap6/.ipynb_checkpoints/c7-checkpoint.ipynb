```python
from pprint import pprint

import numpy as np
import numpy.random as random
import scipy as sp
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
sns.set()
%matplotlib inline

%precision 3
```

```python
x = [1, 2, 3]
y = [10, 1, 4]

plt.figure(figsize=(10, 6))
plt.bar(x, y, align='center', width=0.5)
plt.xticks(x, ['A Class', 'B Class', 'C Class'])
plt.xlabel('Class')
plt.ylabel('Socre')
plt.grid(True)
```

```python
x = [1, 2, 3]
y = [10, 1, 4]

plt.figure(figsize=(10, 6))
plt.barh(x, y, align='center')
plt.yticks(x, ['A Class', 'B Class', 'C Class'])
plt.ylabel('Class')
plt.xlabel('Score')
plt.grid(True)
```

```python
y1 = np.array([30, 10, 40])
y2 = np.array([10, 50, 90])

x = np.arange(len(y1))

w = 0.4

plt.figure(figsize=(10, 6))
plt.bar(x, y1, color='blue', width=w, label='Math first', align='center')
plt.bar(x + w, y2, color='green', width=w, label='Math final', align='center')
plt.legend(loc='best')
plt.xticks(x + w / 2, ['Class A', 'Class B', 'Class C'])
plt.grid(True)
```

```python
height1 = np.array([100, 200, 300, 400, 500])
height2 = np.array([1000, 800, 600, 400, 200])

x = np.array([1, 2, 3, 4, 5])

plt.figure(figsize=(10, 6))
p1 = plt.bar(x, height1, color='blue')
p2 = plt.bar(x, height2, bottom=height1, color='lightblue')
plt.legend((p1[0], p2[0]), ('Class 1', 'Class 2'))

print('p1')
pprint(p1.__dict__)
print('\np2')
pprint(p2.__dict__ )
print()

print('p1[0]:', p1[0])
print('type(p1[0]):', type(p1[0]))
```

```python
labels = ['Frogs', 'Hogs', 'Dogs', 'Logs']
sizes = [15, 30, 45, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)

plt.figure(figsize=(15, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=0)

plt.axis('equal')
```

```python
plt.pie(sorted(sizes, reverse=True), explode=explode, labels=labels, colors=colors, autopct='%1.2f%%', shadow=True, 
        startangle=90, counterclock=False)
plt.axis('equal')
```

```python
N = 25

x = np.random.rand(N)
y = np.random.rand(N)

colors = np.random.rand(N)

area = 10 * np.pi * (15 * np.random.rand(N)) ** 2

plt.figure(figsize=(15, 6))

plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.grid(True)
```

```python
reason_count_series = student_data_math['reason'].value_counts()

labels = reason_count_series.index.values
sizes = reason_count_series.values
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']

plt.figure(figsize=(15, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90, counterclock=False)
plt.axis('equal')
```

```python
higher_count_series = student_data_math['higher'].value_counts()
higher_count_series
```

```python
g3_means = student_data_math.groupby(['higher'])['G3'].mean()
g3_means.sort_values(ascending=False)
```

```python
g3_means.plot(kind='bar')
```
